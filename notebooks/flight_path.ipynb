# **Batted Baseball Flight Path**

The goal of this file is to demonstrate the `homerun.py` algorithm.

```python
import numpy as np
from utils.homerun import HomeRun
```

### [Mike Trout Three-Run Homer](https://baseballsavant.mlb.com/sporty-videos?playId=bcb2ed80-3b54-3038-81f7-41e5d9de52e4)

```python
# Define HomeRun model parameters

# Ball constants 
m = 0.145           # Mass of baseball (kg)
g = 9.807           # Gravity (m/s^2)
c_d = 0.12          # Effective drag coefficient scaled for raw ODE
c_l = 0.10          # Effective lift coefficient scaled for raw ODE
r = 0.0366          # Radius of baseball (meters)

# Mike Trout Statcast Data: August 8, 2025
rho = 1.200                          # Air density approximation for Anaheim, CA (kg/m^3)
s_mph = 107.5                        # Exit speed (mph)
s = s_mph * 0.44704                  # Exit speed (m/s)
theta = 28.0                         # Launch angle (degrees) ESTIMATE
phi = 55.0                           # Spray angle (degrees)  ESTIMATE                     
h = 0.65                             # Contact height (meters)
omega = np.array([2000 * np.sin(np.radians(112.5)), -2000 * np.cos(np.radians(112.5)), 0])      # Backspin vector (RPM) ESTIMATE
```

```python
# Instantiate HomeRun object

hr = HomeRun(
    m = m,
    g = g,
    rho = rho,
    c_d = c_d,
    c_l = c_l,
    r = r,
    h = h,
    s = s,
    theta = theta,
    phi = phi,
    omega = omega
)
```

```python
# Coordinate System

ft_to_m = 0.3048

# Angle Stadium distances to outfield wall from home base
phi_sectors_R = [0.0, 11.0, 22.0, 32.0, 45.0, 58.0, 90.0]  
distances_ft  = [
    350.0,   # RF line to shallow RF
    365.0,   # Shallow RF to shallow RC
    370.0,   # Shallow RC to RC gap
    396.0,   # RC gap to CF
    390.0,   # CF to LC gap
    347.0,   # LC gap to LF line
]
distances = [d * ft_to_m for d in distances_ft]

# Angle Stadium heights of outfield wall
phi_sectors_H = [0.0, 11.0, 22.0, 32.0, 45.0, 58.0, 90.0]
heights_ft = [
    5.0,    # RF corner 
    8.0,    # Shallow RF to shallow RC
    8.0,    # Shallow RC to RC gap
    8.0,    # RC gap to CF
    8.0,    # CF to LC gap
    8.0,    # LC gap to LF line
]
heights = [ht * ft_to_m for ht in heights_ft]

hr.set_R(phi_sectors_R, distances)
hr.set_H(phi_sectors_H, heights)
```

```python
# Solve trajectory

dt = 0.01
maxiter = 10_000

hr.solve_trajectory(dt, maxiter)
```

```python
# Classify batted ball

is_hr = hr.homerun_classifier()
print(f"Home Run at Angels Stadium: {is_hr}")
```
